**Each stage of construction has a separate process defined. All processes are the same, just with different quantities of time, resources and money**

```python
def rebuild_plinth(env, inspection_program, insurance_program, nra_program, loan_program, 
                 assessment_program, permit_program, rebuild_program, entity):
        
    #inspect house
    yield env.process(inspection_program.process(entity.property, entity))
    
    #entity.property refers to the individual building in structures.py


    if entity.property.damage_state == 'None' :
        # household not eligible for funding
        entity.story.append('Household has no damage')
        
    elif entity.property.damage_state == 'Slight' :
        # household not eligible for funding
        entity.story.append('Household not eligible for funding so must fund repairs themselves')
        # could yield a process where they do this
        
    elif entity.property.damage_state == 'Moderate' or 'Extensive' or 'Complete':
        yield env.process(permit_program.process(entity.property, entity))
            # create funding search policy
        yield env.process(plinth_funding_search.policy(insurance_prog, nra_prog, loan_prog, entity))
        yield env.timeout(190)
        yield env.process(rebuild_program.process(entity.property, entity))
            # rebuild program checks for materials and labour
        #reset inspection
        entity.property.inspected = False
        
```

```python
def rebuild_walls(env, inspection_program, insurance_program, nra_program, loan_program, 
                 assessment_program, permit_program, rebuild_program, entity):
   
        #inspect house
        yield env.process(inspection_program.process(entity.property, entity))
    
        #entity.property refers to the individual building in structures.py

        if entity.property.damage_state == 'None' :
            # household not eligible for funding
            entity.story.append('Household has no damage')
            # could yield a process where they do this
        
        elif entity.property.damage_state == 'Slight' :
            # household not eligible for funding
            entity.story.append('Household not part of scheme')
        # can't combine these two if statements for 'none' and 'slight' for some reason
        
        elif entity.property.damage_state == 'Moderate' or 'Extensive' or 'Complete':
            #yield env.process(permit_program.process(entity.property, entity))
                # create funding search policy
            yield env.process(walls_funding_search.policy(insurance_prog, nra_prog, loan_prog, entity))
            yield env.timeout(180)
            yield env.process(rebuild_program.process(entity.property, entity))
            # rebuild program checks for materials and labour

            
```

```python
def rebuild_roof(env, inspection_program, insurance_program, nra_program, loan_program, 
                 assessment_program, permit_program, rebuild_program, entity):
 
        #inspect house
        yield env.process(inspection_program.process(entity.property, entity))
    
        #entity.property refers to the individual building in structures.py

        if entity.property.damage_state == 'None' :
            # household not eligible for funding
            entity.story.append('Household has no damage')
        
        elif entity.property.damage_state == 'Slight' :
            # household not eligible for funding
            entity.story.append('Household not part of scheme')
        
        elif entity.property.damage_state == 'Moderate' or 'Extensive' or 'Complete':
            #yield env.process(permit_program.process(entity.property, entity))
                # create funding search policy
            yield env.process(roof_funding_search.policy(insurance_prog, nra_prog, loan_prog, entity))
            yield env.timeout(25)
            yield env.process(rebuild_program.process(entity.property, entity))
            # rebuild program checks for materials and labour

```

## Combine processes to work sequentially after each stage

```python
def combinedprocess(env,entity):
    #for i in range(len(owners)):
    if entity.property.new_plinth == False:
        yield env.process(rebuild_plinth(env, inspection_prog, insurance_prog, nra_prog, loan_prog, 
                                    assessment_prog, permit_prog, rebuild_plinth_prog, entity))
    if entity.property.new_plinth == True:
        yield env.process(rebuild_walls(env, inspection_prog, insurance_prog, nra_prog, loan_prog, 
                                    assessment_prog, permit_prog, rebuild_walls_prog, entity))
    if entity.property.new_walls == True:
        yield env.process(rebuild_roof(env, inspection_prog, insurance_prog, nra_prog, loan_prog, 
                                    assessment_prog, permit_prog, rebuild_roof_prog, entity))

            
for i in range(len(owners)):
    env.process(combinedprocess(env,owners[i]))
```

** Replenishes material stocks over time **

```python
def materialprocess(env):
    for i in range(100000000):
        monthly_percent = 0.0002/12
        yield env.timeout(30)
        yield stone.put(10000*monthly_percent)  
        yield brick.put(430000000*monthly_percent)
        yield timber.put(550000*monthly_percent)
        yield rebar.put(10000**monthly_percent) 
        yield cement.put(70000000*monthly_percent) 
        yield cgi.put(14500000*monthly_percent)
        yield aggregate.put(830000*monthly_percent) 
        yield sand.put(2610000*monthly_percent)
        yield skilled.release(1) #this doesn't seem to do anything
        i +=1
        #effect of embargo stopping increase in material and reducing timber
        #if i == 4:
            #yield timber.get(550000*0.0001)
            #yield env.timeout(135)


env.process(materialprocess(env))

def incomeprocess(env,entity):
        for i in range(100000000):
            yield env.timeout(7)
            #add 40% income to money to build each week
            entity.money_to_rebuild += (entity.income)*0.4*6
for i in range(len(owners)):
    env.process(incomeprocess(env,owners[i]))
```

**Collects material quantities over time and sets up length of simulation with a timestep of 1 month (30 days)** 

```python
# yields quantities over time (eg materials)  -time set to 2 years
time = list(range(1, 691, 30)) #originally 751, 30
i=0
stn=[]
cem=[]
reb=[]
bri=[]
agg=[]
san=[]
tim=[]
count=[]
money=[]
inspect_staff=[]
permit_staff=[]
#checks money for owner 0, graph output doesn't show input and output for roof as it was constructed and paid for within 1 month 
house = owners[0]
while i < 23:
    env.run(until=time[i])
    sstone = [stone.level]
    stn.extend(sstone)
    scement=[cement.level]
    cem.extend(scement)
    srebar=[rebar.level]
    reb.extend(srebar)
    sbrick=[brick.level]
    bri.extend(sbrick)
    saggregate=[aggregate.level]
    agg.extend(saggregate)
    ssand=[sand.level]
    san.extend(ssand)
    stimber=[timber.level]
    tim.extend(stimber)
    cash=[house.money_to_rebuild]
    money.extend(cash)
    
    sskilled=[skilled.count]
    count.extend(sskilled)
    
    inspect=[inspection_prog.staff.count]
    inspect_staff.extend(inspect)
    
    permitt=[permit_prog.staff.count]
    permit_staff.extend(permitt)
    i += 1
```

### Statistical output used to compare scenarios at a community level

```python
num_damaged=0
num_slight=0
num_new_plinth=0
num_new_walls=0
num_new_roof=0
tot_plinth_time =0
tot_walls_time=0
tot_roof_time=0
time_till_plinth =0
time_till_walls=0
time_till_roof=0
ave_plinth_time =0
ave_walls_time = 0
ave_roof_time =0
ave_time_till_plinth=0
ave_time_till_walls = 0
ave_time_till_roof =0
tot_debt =0
num_first_inst = 0
num_second_inst = 0
num_third_inst = 0
for household in owners:
    if household.residence.damage_state!= 'None': num_damaged +=1
    if household.residence.damage_state== 'Slight': num_slight +=1
    if household.residence.new_plinth == True: 
        num_new_plinth +=1
        tot_plinth_time += household.residence.plinth_time
        time_till_plinth += household.residence.finished_plinth
    if household.residence.new_walls == True: 
        num_new_walls +=1
        tot_walls_time += household.residence.walls_time
        time_till_walls += household.residence.finished_walls
    if household.residence.new_roof == True: 
        num_new_roof +=1
        tot_roof_time += household.residence.roof_time
        time_till_roof += household.residence.finished_roof
    if household.residence.first_inst == True:
        num_first_inst += 1
    if household.residence.second_inst == True:
        num_second_inst += 1
    if household.residence.third_inst == True:
        num_third_inst += 1
    tot_debt += household.debt
ave_debt = tot_debt/len(owners)
ave_plinth_time = (tot_plinth_time)/(num_new_plinth)
if num_new_walls != 0:
    ave_walls_time = (tot_walls_time)/(num_new_walls)
    ave_time_till_walls = (time_till_walls)/(num_new_walls)
if num_new_roof != 0:
    ave_roof_time = (tot_roof_time)/(num_new_roof)
    ave_time_till_roof = (time_till_roof)/(num_new_roof)
ave_time_till_plinth = (time_till_plinth)/(num_new_plinth)
print('{0} out of {1} households suffered damage to their homes.\n'.format(num_damaged,len(owners)),
      '{0} out of {1} households suffered slight damage so did not receive a grant.\n'.format(num_slight,len(owners)),
      '{0} out of {1} households received the first tranche of funding.\n'.format(num_first_inst,len(owners)),
      '{0} out of {1} owners rebuilt to the plinth level.\n'.format(num_new_plinth, len(owners)),
      '{0} out of {1} households received the second tranche of funding.\n'.format(num_second_inst,len(owners)),
      '{0} out of {1} owners rebuilt to the walls level.\n'.format(num_new_walls, len(owners)),
      '{0} out of {1} households received the third tranche of funding.\n'.format(num_third_inst,len(owners)),
      '{0} out of {1} owners completed reconstruction to the roof level.\n'.format(num_new_roof, len(owners)),
      'Average time taken to rebuild plinth: {0:.0f} days.\n'.format(ave_plinth_time),
      'Average time taken to rebuild superstructure: {0:.0f} days.\n'.format(ave_walls_time),
      'Average time taken to rebuild roof: {0:.0f} days.\n'.format(ave_roof_time),
      'Average time taken to rebuild to plinth level: {0:.0f} days.\n'.format(ave_time_till_plinth),
      'Average time taken to rebuild to superstructure level: {0:.0f} days.\n'.format(ave_time_till_walls),
      'Average time taken to complete reconstruction process: {0:.0f} days.\n'.format(ave_time_till_roof),
      'Average debt: Rs {0:.0f}'.format(ave_debt)
     )
```
